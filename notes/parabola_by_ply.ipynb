```python
#!pip install plotly
import plotly.graph_objects as go
import numpy as np
# 定义抛物线参数
a = 1
b = 0
c = 0

# 生成x的值
x = np.linspace(-10, 10, 400)

# 计算对应的y值
y = a * x**2 + b * x + c

# 创建图形
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='y = ax^2'))
fig.update_layout(title='Parabola',
                  xaxis_title='x',
                  yaxis_title='y')
fig.show()
```
