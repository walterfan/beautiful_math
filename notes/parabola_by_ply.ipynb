```python
#!pip install plotly
import plotly.graph_objects as go
import numpy as np
# 
a = 1
b = 0
c = 0

# x
x = np.linspace(-10, 10, 400)

# y
y = a * x**2 + b * x + c

# 
fig = go.Figure(data=go.Scatter(x=x, y=y, mode='lines', name='y = ax^2'))
fig.update_layout(title='Parabola',
                  xaxis_title='x',
                  yaxis_title='y')
fig.show()
```
